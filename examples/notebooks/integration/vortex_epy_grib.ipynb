## Downloading the input file

For this example, we will use a GRIB file produced by the operational suite using the Arome model over the French West Indies.

```python
rhS = vortex.toolbox.rload(local='epytest_[geometry:area]_[term]',
                           kind='gridpoint',
                           format='grib',
                           geometry='antil0025',
                           origin='historic',
                           experiment='OPER',
                           block='forecast',
                           date='2016100100',
                           term='0',
                           cutoff='production',
                           namespace='vortex.archive.fr',
                           vapp='arome',
                           vconf='antilles',
                           model='[vapp]',
                           nativefmt='[format]')
```

```python
for rh in rhS:
    rh.get()
```

## Using Epygram to inspect the GRIB file

First, we need to make sure that the epygram package is loaded.

```python
from common.util import usepygram
```

Once the epygram package is availlable, the ResourceHandler's Content object is abled to return a proper epygram object that allows to access the data within the GRIB file. 

```python
rh0 = rhS[0]
print 'Number of messages:', rh0.contents.data.messages_number
```

We can retrieve a specific field, compute statistics and plot a map !

```python
field = rh0.contents.data.readfield(dict(shortName='t', typeOfLevel='isobaricInhPa', level=850))
field.stats()
```

```python
fig = field.plotfield(title="A great plot !")
```
